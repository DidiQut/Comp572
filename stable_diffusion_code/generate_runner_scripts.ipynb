# Create Command txt for generating panda picture

```python
artist_names = ['Vincent van Gogh', 'Claude Monet','Pablo Picasso','Sandro Botticelli',
                'Leonardo da Vinci','Albrecht Dürer','Michelangelo', 'Raphael',
                'Caravaggio', 'Peter Paul Rubens','Artemisia Gentileschi','Gian Lorenzo Bernini',
                'Diego Velázquez', 'Rembrandt', 'Jan Vermeer', 'Edo Period', 'Katsushika Hokusai',
                'Utagawa Hiroshige', 'Eugène Delacroix', 'Édouard Manet', 'Edgar Degas', 
                'Mary Cassatt', 'Paul Cézanne', 'Paul Gauguin', 'Henri Matisse',
                'Gustav Klimt', 'Amadeo Modigliani', 'Diego Rivera', 'René Magritte',
                'Salvador Dalí', 'Frida Kahlo', 'Mark Rothko', 'Jackson Pollock']
seed_list = list(range(5000))
```

```python
len(artist_names)
```

```python
%%time
command_list = []
for curr_seed in seed_list:
    curr_artist = np.random.choice(artist_names,1, p=[0.3, 0.1, 0.1] + [0.5/30]*30)[0]
    command = """python scripts/txt2img.py --prompt "panda by {}" --plms --ckpt sd-v1-4.ckpt --skip_grid --n_samples 1 --n_iter 1 --ddim_steps 150 --H 448 --W 448 --seed {} --outdir "outputs/txt2img-samples/{}_{}" """.format(curr_artist, curr_seed, curr_seed, curr_artist).strip()
    command_list.append(command)
```

```python
with open("command_list.txt", 'w', encoding='utf-8') as file:
    file.write('\n'.join(command_list))
```
